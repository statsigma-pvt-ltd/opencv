```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
import community
import pandas as pd
```

```python
G_fb = nx.read_edgelist("C:/Users/akhil/Downloads/facebook_data/facebook_combined/subfiletw.txt", create_using = nx.Graph(), nodetype = int)
```

```python
#Quick snapshot of the Network
print(nx.info(G_fb))
```

```python
spring_pos = nx.spring_layout(G_fb)
```

```python
plt.figure(figsize=(18,18))
plt.axis("off")
nx.draw_networkx(G_fb, pos = spring_pos, with_labels = False, node_size = 35)
plt.savefig('C:/Users/akhil/Downloads/facebook_data/facebook_combined/nodes_edges.png')
```

```python
parts = community.best_partition(G_fb)
values = [parts.get(node) for node in G_fb.nodes()]
```

```python
plt.figure(figsize=(18,18))
plt.axis("off")
var=nx.draw_networkx(G_fb, pos = spring_pos, cmap = plt.get_cmap("jet"), node_color = values, node_size = 35, with_labels = True)
plt.savefig('C:/Users/akhil/Downloads/facebook_data/facebook_combined/clusters_community.png')
```
